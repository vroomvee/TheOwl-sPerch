```python


def parse_float_scores(text):
  """
  This function parses text and extracts floating-point scores 
  in the format 'X.X (out of Y)' or 'X.X rating out of Y'.

  Args:
      text: A string containing the text to parse.

  Returns:
      A list of extracted float scores, or an empty list if none are found.
  """
  scores = []
  # Improved regular expression to avoid matching "out"
  pattern = r"\d+\.\d+(?!out)"  # Match digits followed by a decimal, not followed by "out"
  matches = re.findall(pattern, text, re.MULTILINE)
  for match in matches:
    scores.append(float(match))
  return scores

# # Sample text
# text = """Tejaswini Swaroop's projects demonstrate unique applications of machine learning, particularly in the development of a movie recommending system. The candidate's ability to utilize various Python libraries and implement data preprocessing techniques such as stemming and vectorization showcases their technical proficiency. Additionally, their experience as a Technology and Fashion Journalist exhibits their ability to communicate complex technical concepts effectively. These qualities align well with the requirements of an ML engineer role, making them a suitable candidate.
# 1.8
# 0.0
# 1.5 rating out of 2
# and just return the float scores = [1.8,1.8,0.0,1.5]"""

# # Extract scores from the text
# extracted_scores = parse_float_scores(text)

# # Print the extracted scores
# print(extracted_scores)
```
